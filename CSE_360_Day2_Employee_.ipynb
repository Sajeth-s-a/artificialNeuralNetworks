```python
#import the libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv("/content/drive/MyDrive/CSE 360 - Deep Learning/Datasets/HR_comma_sep.csv")
data.head(5)
```

```python
print(data.shape)
```

```python
df1=['sales','salary']
df_final = pd.get_dummies(data,columns = df1,drop_first = True)
df_final.head(3)
```

```python
from sklearn.model_selection import train_test_split
X = df_final.drop(['left'],axis = 1).values
y = df_final['left'].values
print(X.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

```python
#Build the model
model = Sequential()
model.add(Dense(128,kernel_initializer="uniform",input_dim = 18,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,kernel_initializer="uniform",activation='sigmoid'))
```

```python
#model compile
model.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
#fit the model
model.fit(X_train,y_train,epochs = 100,batch_size = 32)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test , verbose=2)

print('\nTest accuracy:', test_acc)
print('\nTest Loss', test_loss)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred1 = model.predict(x_test)
y_pred1 = np.argmax(y_pred1,axis = 1)
cm = confusion_matrix(y_test,y_pred1)
print(cm)
print(accuracy_score(y_test,y_pred1))
```
